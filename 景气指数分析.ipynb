```python
from matplotlib import pyplot as plt
import matplotlib
import pandas as pd

font = { 'family' : 'SimSun',
        'weight' : 'bold',
         'size'   : '20'}
matplotlib.rc('font',**font)
```

```python
my_data = pd.read_csv('样例\景气指数_加盐.csv',encoding='utf-8')
```

```python
my_data['起飞星期']
```

```python
new_data=my_data.groupby('起飞年')['景气指数'].mean()
```

```python
new_data
```

```python
plt.figure(figsize = (6, 6))
# 设置x轴和y轴的值
x = new_data.index
y = new_data.values
# 绘制折线图
plt.plot(x, y, linewidth = 2, color = 'r', marker = 'o',
         markerfacecolor = 'blue', markersize = 8)
plt.xticks(size = 12)
plt.yticks(size = 12)
# 添加x轴和y轴的标签
plt.xlabel('起飞年', fontdict = {'size': 15})
plt.ylabel('景气指数', fontdict = {'size': 12})
plt.title('各年份景气指数变化图', fontsize = 15)
# 保存图形
plt.savefig('样例\市场景气指数.jpg')
```

```python
my_data['起飞时间']=my_data['起飞年']+my_data['起飞星期']%12/100
```

```python
new_data = my_data.groupby('起飞时间')['景气指数'].mean()
```

```python
my_data
```

```python
plt.figure(figsize = (15, 6))
# 设置x轴和y轴的值
x = new_data.index
y = new_data.values
# 绘制折线图
plt.plot(x, y, linewidth = 2, color = 'r', marker = 'o',
         markerfacecolor = 'blue', markersize = 8)
plt.xticks(size = 12)
plt.yticks(size = 12)
# 添加x轴和y轴的标签
plt.xlabel('起飞年', fontdict = {'size': 15})
plt.ylabel('景气指数', fontdict = {'size': 12})
plt.title('各年份景气指数变化图', fontsize = 15)
# 保存图形
plt.savefig('样例\市场景气指数.jpg')
```

```python

order_number = my_data['起飞年'].value_counts()/my_data['起飞年'].value_counts().sum()
```

```python
order_number
```

```python
x=order_number
```

```python
# 生成画布，并设置画布的大小
plt.figure(figsize = (6, 6))
# 设置textprops参数值
textprops = {'fontsize': 10, 'color': 'black'}
# 设置explode参数值
explode = [0, 0, 0.05, 0]
# 绘制饼图（包含调整图形样式）
plt.pie(x, autopct = '%.2f%%', textprops = textprops, explode = explode)
```

```python
# 生成画布，并设置画布的大小
plt.figure(figsize = (6, 6))
# 设置textprops参数值
x=order_number.values
textprops={'fontsize': 10, 'color': 'black'}
# 设置explode参数值
explode=[0, 0.05, 0.05, 0.05]
# 绘制饼图（包括调整图形样式）
plt.pie(x, autopct = '%.2f%%', textprops = textprops, explode = explode)
# 设置图形的标题
plt.title('起飞年', fontsize = 15)
# 设置图例
plt.legend(labels = order_number.index)
# 保存图形
plt.savefig('样例\起飞分布.jpg')
```
